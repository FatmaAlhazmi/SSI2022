# Stability Comparison for Various Schemes 

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# the resolution
ny = 200
nx = 200

# creat arrays that store x and y points
y = np.linspace(-5, 1.5, ny)
x = np.linspace(-5, 3.5, nx)

# creat meshgrid
Y, X = np.meshgrid(x, y)

# complex numbers' space
Z = X + 1j*Y

# the remaining terms from Taylor expansion of Euler's and its norm
lambda1 = 1 + Z
norm1 = np.real(lambda1*lambda1.conj())

# same for RK2
lambda2 = 1 + Z + Z**2/2.
norm2 = np.real(lambda2*lambda2.conj())

# same for RK4
lambda3 = 1 + Z + Z**2/2. + Z**3/6. + Z**4/24.
norm3 = np.real(lambda3*lambda3.conj())
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.contour(X, Y, norm1, levels=[1], colors='b')
ax.contour(X, Y, norm2, levels=[1], colors='y')
ax.contour(X, Y, norm3, levels=[1], colors='r')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('center')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)



ax.text(-1, 1.1, r'Euler', fontsize=14, horizontalalignment='center')
ax.text(-1, 1.85, r'RK2', fontsize=14, horizontalalignment='center')
ax.text(-2.05, 2.05, r'RK4', fontsize=14, horizontalalignment='center')
```
