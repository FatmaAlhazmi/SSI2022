```python
# import
import numpy as np
import matplotlib.pyplot as plt

# variables
dt = 0.1
t = np.linspace(-2,4)
f = np.sin(t)

# exact derivative
dfdt = np.cos(t)

# forward difference
dfdtFD = (np.sin(t+dt)-np.sin(t))/dt;

# plotting
plt.plot(t,f, 'k');
plt.plot(t,dfdt, '--k');
plt.plot(t,dfdtFD, 'b', 1.2);
plt.legend(['Function','Exact Derivative', 'Forward'])

plt.grid()
plt.show()

# Compute max error  
max_error = max(abs(dfdt - dfdtFD))
print(max_error)
```

```python
# backward difference 
dfdtBD = (np.sin(t)-np.sin(t-dt))/dt;

# plotting
plt.plot(t,f, 'k');
plt.plot(t,dfdt, '--k');
plt.plot(t,dfdtFD, 'g', 1.2);
plt.legend(['Function','Exact Derivative', 'Backward'])

plt.grid()
plt.show()

# Compute max error  
max_error = max(abs(dfdt - dfdtBD))
print(max_error)
```

```python
# central difference 
dfdtCD = (np.sin(t+dt)-np.sin(t-dt))/(2*dt);

# plotting
plt.plot(t,f, 'k');
plt.plot(t,dfdt, '--k');
plt.plot(t,dfdtCD, 'r', 1.2);
plt.legend(['Function','Exact Derivative','Central'])

plt.grid()
plt.show()

# Compute max error  
max_error = max(abs(dfdt - dfdtCD))
print(max_error)
```

## The three formulas combined in one figure:

```python
# import
import numpy as np
import matplotlib.pyplot as plt

# variables
dt = 0.1
t = np.linspace(-2,4)
f = np.sin(t)


# exact derivative
dfdt = np.cos(t)

# forward difference
dfdtFD = (np.sin(t+dt)-np.sin(t))/dt;

# backward difference 
dfdtBD = (np.sin(t)-np.sin(t-dt))/dt;

# central difference 
dfdtCD = (np.sin(t+dt)-np.sin(t-dt))/(2*dt);

# plotting
plt.plot(t,f, '--k');
plt.plot(t,dfdt, 'k');
plt.plot(t,dfdtFD, 'b', 1.2);
plt.plot(t,dfdtBD, 'g', 1.2);
plt.plot(t,dfdtCD, 'r', 1.2);
plt.legend(['Function','Exact Derivative','Forward','Backward','Central'])

plt.grid()
plt.show()
```

## Note that if we took a smaller dt, we will have more accurate approximations. For instance, let dt = 0.01 instead of 0.1:

```python
# import
import numpy as np
import matplotlib.pyplot as plt

# variables
dt = 0.01
t = np.linspace(-2,4)
f = np.sin(t)


# exact derivative
dfdt = np.cos(t)

# forward difference
dfdtFD = (np.sin(t+dt)-np.sin(t))/dt;

# backward difference 
dfdtBD = (np.sin(t)-np.sin(t-dt))/dt;

# central difference 
dfdtCD = (np.sin(t+dt)-np.sin(t-dt))/(2*dt);

# plotting
plt.plot(t,f, '--k');
plt.plot(t,dfdt, 'k');
plt.plot(t,dfdtFD, 'b', 1.2);
plt.plot(t,dfdtBD, 'g', 1.2);
plt.plot(t,dfdtCD, 'r', 1.2);
plt.legend(['Function','Exact Derivative','Forward','Backward','Central'])

plt.grid()
plt.show()
```
