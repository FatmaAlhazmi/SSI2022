# Coding Heat Equation in 3D

```python
# Enable interactive plot
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import seaborn as sns
sns.set()


gamma = 1/4
x = np.linspace(0,1,101)
y = np.linspace(0,1,101)
X,Y = np.meshgrid(x,y)
u = np.zeros((101,101))
u[40,50] = 100
T = 400
fig = plt.figure()
ax = plt.axes(projection='3d')
ax._axis3don = False
c = ax.contourf3D(X,Y,u,1000, cmap='pink')

def ans(f):
    global u, c

    #Time is not defined as the animation will do the loop for T times.
    for j in range(100):
        for i in range(100):
            u[i, j] = gamma * (u[i+1][j] + u[i-1][j] + u[i][j+1] + u[i][j-1] - 4*u[i][j]) + u[i][j]
    for xx in c.collections:
        xx.remove()
    c = ax.contourf3D(X,Y,u,1000, cmap='pink')
    return c

anim = animation.FuncAnimation(fig, ans, frames=T)
anim.save("3dheat_equation_solution.gif")
plt.show()

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()    
```
