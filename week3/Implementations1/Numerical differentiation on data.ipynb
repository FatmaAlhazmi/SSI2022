```python
# import
import numpy as np
import matplotlib.pyplot as plt

# variables
x = np.arange(0.1,3, 0.1)
f = np.sin(x)

plt.plot(x,f,'k')
plt.plot(x,f,'rx')

plt.show()
```

### The red x's are the data points that we have been given, while the black curve is the true function that these points have been sampled from.

```python
# dx = x[1] - x[0]
n = len(f)
dx = x[2]-x[1]

dfdx = np.zeros((n,1))

dfdx[0] = (f[1]-f[0])/dx # finite difference method for the first point

for i in range(1, n-1):
    dfdx[i] = (f[i+1]-f[i-1])/(x[i+1]-x[i-1]) # central difference method
    
dfdx[n-1] = (f[n-1]-f[n-2])/(x[n-1]-x[n-2]) # backward difference method for the last point
 
    
plt.plot(x,np.cos(x),'k') # the real derivative of sin(x)
plt.plot(x,dfdx,'rx') # our approximation 
```

### The black curve is the true derivative "cosine", while the red x's are the approximations. Note that its a little more accurate in the center because we used CDM, as we couldn't use it for the left and right most points.
