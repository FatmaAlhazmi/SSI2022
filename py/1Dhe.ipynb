# Coding Heat Equation in 1D

```python
# import numpy and matplotlib
import numpy as np
from matplotlib import pyplot as plt

#Create a numpy linspace array
L = 100
gamma = 1/4

x = np.linspace(0, 1, L)
u = np.zeros(L)

#Initial Condition
u[L//2] = 1  

#Boundary Conditions are not necessary 

# time loop

for t in range(100):
    for i in range(1, L-1):
        
        u[i] = gamma * (u[i+1] + u[i-1] - 2*u[i]) + u[i]
        
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x,u, linewidth=5, c=u, cmap='pink')
plt.show()
```
